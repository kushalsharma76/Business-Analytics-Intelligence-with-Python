```python
# #Import Packages
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#Load iris dataset
iris = datasets.load_iris()
print(iris.data)
print(iris.target)
```

```python
#target & predictors
#input
X = iris.data[:,:2]
print(X)
#output
Y = iris.target
print(Y)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y, cmap = "gist_rainbow")
#The colormap is a dictionary which maps numbers to colors. Matplotlib provides many built-in colormaps.
plt.xlabel('Spea1 Length', fontsize=14)
plt.ylabel('Sepal Width', fontsize=14)
plt.title("Iris Data", fontsize=20)
```

```python
km = KMeans(n_clusters = 3)
```

```python
# km1 = KMeans(n_clusters = 3, n_jobs = 1)
```

```python
km.fit(X)
# km1.fit(X)
```

```python
print(km.cluster_centers_)
# print(km1.cluster_centers_)
```

```python
new_labels = km.labels_
print(new_labels)
# new_labels1 = km1.labels_
```

```python
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=Y, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```
